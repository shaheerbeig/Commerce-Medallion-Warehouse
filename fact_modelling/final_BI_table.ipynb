```python
%sql
create or replace view ecommerce.gold.gold_dahboard as
select 
o.*,d.quarter,d.Month,d.is_weekend, p.category_name,p.category_code,p.brand_name, p.brand_code  from ecommerce.gold.order_base_table o join ecommerce.gold.date_dim d on o.dt = d.date join ecommerce.gold.product_dim p on o.product_id = p.product_id  
```
